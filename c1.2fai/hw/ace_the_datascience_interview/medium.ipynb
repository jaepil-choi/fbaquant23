1st try: 

- 한 번에 3개씩 iterative

```python
test1 = [3, 5, 2, 4, 1]
```

```python
def get_peaks(li):
    peaks = []
    for one, two, three in zip(li, li[1:], li[2:]):
        if one < two > three:
            peaks.append(two)
    
    return peaks
```

```python
get_peaks(test1)
```

2nd try: 
- 2번째가 peak이면 1번째, 3번째는 분명히 peak이 아니다. 검사할 필요가 없다. 
- 1st try에선 모든걸 iterative하게 보고 있어서 비효율적이다. 

```python
def get_peaks2(li):
    peaks = []

    skip_idx = None

    for idx, val in enumerate(li):
        if idx == 0:
            continue
            
        if idx == len(li) - 1:
            break

        if idx == skip_idx:
            continue
        
        if li[idx - 1] < val > li[idx + 1]:
            peaks.append(val)
            skip_idx = idx + 1
    
    return peaks
```

```python
get_peaks2(test1)
```

compare two algorithms

```python
length = 100000
test_big = np.random.randint(1, 100, length)
```

```python
%%timeit -r 10 -n 100

get_peaks2(test_big)
```

```python
%%timeit -r 10 -n 100

get_peaks(test_big)
```
