1st try:

```python
test1 = [3, 5, 2, 4, 1]
```

```python
def get_peaks(li):
    peaks = []
    for one, two, three in zip(li, li[1:], li[2:]):
        if one < two > three:
            peaks.append(two)
    
    return peaks
```

```python
get_peaks(test1)
```

2nd try: 
- 2 peak 1, 3  peak .   . 
- 1st try  iterative   . 

```python
def get_peaks2(li):
    peaks = []

    skip_idx = None

    for idx, val in enumerate(li):
        if idx == 0:
            continue
            
        if idx == len(li) - 1:
            break

        if idx == skip_idx:
            continue
        
        if li[idx - 1] < val > li[idx + 1]:
            peaks.append(val)
            skip_idx = idx + 1
    
    return peaks
```

```python
get_peaks2(test1)
```

compare two algorithms

```python
length = 100000
test_big = np.random.randint(1, 100, length)
```

```python
%%timeit -r 10 -n 100

get_peaks2(test_big)
```

```python
%%timeit -r 10 -n 100

get_peaks(test_big)
```
