```python
class Bond:
    def __init__(self, coupon_r, years, fv, price) -> None:
        self.coupon_r = coupon_r
        self.years = years
        self.fv = fv
        self.price = price
    
    def get_sympy_expr(self):
        expr = self.price - self.fv/(1+r)**self.years - sympy.Sum(
            self.coupon_r * self.fv / (1+r)**k, 
            (k, 1, self.years)
            )

        return expr
    
    def get_discount_r(self):
        
        return sympy.nsolve(self.get_sympy_expr(), 0)
```

```python
cur_shrt_r = 0.0

bond_A = Bond(0.04, 10, 1000, 950)
bond_B = Bond(0.04 + cur_shrt_r, 10, 1000, 1100)
bond_C = Bond(0.08 - cur_shrt_r, 10, 1000, 900)
```
