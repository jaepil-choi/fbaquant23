```python
low = 0
high = 1
trials = int(1e+6)

X1 = np.random.uniform(low, high, size=trials)
X2 = np.random.uniform(low, high, size=trials)

Y = np.minimum(X1, X2)
Z = np.maximum(X1, X2)
```

```python
y = np.random.uniform(low, high, trials)
z = np.random.uniform(low, high, trials)
```

### (a)

P(Y>=y | Z<=z), s.t y, z sampled from [0, 1]

A: Y>=y
B: Z<=z

= P(A|B) = P(A intersect B) / P(B)

```python
# P(B)

def get_B(Z, z):
    return Z <= z
```

```python
# P(A)

def get_A(Y, y):
    return Y >= y
```

```python
# P(A intersect B)

def get_AandB(A, B):
    return A * B
```

```python
A = get_A(Y, y)
B = get_B(Z, z)

sum(get_AandB(A, B)) / sum(get_B(Z, z))
```

## Problem 5

https://leetcode.com/problems/move-zeroes/

```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        zero_count = 0
        i = 0
        while i < len(nums):
            if nums[i] == 0:
                nums.pop(i)
                zero_count += 1
                continue

            i += 1
        
        nums.extend([0] * zero_count)

        return 
```

## Problem 6

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        str_matrix = [list(s) for s in strs]
        min_str_length = min([len(s) for s in strs])

        def check_all_same(list_of_char: list) -> bool:
            uniq = set(list_of_char)
            if len(uniq) == 1:
                return uniq.pop()
            else:
                return None
        
        lc_prefix = ""
        for i in range(min_str_length):
            list_of_chars = [s_list[i] for s_list in str_matrix]
            common = check_all_same(list_of_chars)
            
            if common:
                lc_prefix += common
            else:
                break

        return lc_prefix
```
