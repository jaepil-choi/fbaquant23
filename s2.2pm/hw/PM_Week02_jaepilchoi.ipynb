# PM hw 02

최재필

Portfolio Selection - Harry Markowitz (1952) 논문 요약



- 포트폴리오 선택의 단계:
    1. 자산의 기대 수익률, 변동성에 대한 관찰(=미래에 대한 믿음)
    2. 미래의 기대수익률 등이 주어질 때 최적 포트폴리오 선택 -- 논문의 핵심
- 전제: diversified portfolio > non-diversified portfolio 
    - 단순히 discounted return을 maximize하는 것은 틀림
    - maximum return & minimum variance 를 동시에 추구해야. (E-V rule)
        - 하지만 return끼리 correlation이 있기 때문에 diversification으로 완전히 variance를 없앨 순 없음. 
- 변수 정의
    - `Y` : $ y_1, y_2, y_3, ... $ 의 값들을 가질 수 있는 확률 변수
    - `E` : $ E = p_1y_1 + p_2y_2 + ... + p_Ny_N $ 즉, expected value (mean)
        - 높을수록 좋음
    - `V` : $ V = p_1(y_1 - E)^2 + p_2(y_2 - E)^2 + ... + p_N(y_N - E)^2  $ 즉, variance
        - 낮을수록 좋음
    - `R` : $ R = a_1*R_1 + a_2*R_2 + ... + a_N*R_N $ 라는 확률 변수 (=기대 수익률의 linear combination)
        - `X`: `a` 대신 쓰는 가중치 (투자금액) - `X`는 투자자가 설정하는 것이므로 random variable이 아님. 
        - `E(R)`: 포트폴리오 기대수익률과
        - `Cov(R_1, R_2)`: 공분산을 구할 수 있음
- E-V maxim (=MVO) 그래프
    - 
        - isomean과 isovariance = 등고선
    - 
        - isomean line
            - 원점이랑 멀리 있을수록 좋음. `E`(mean)이 높은 것. 
        - isovariance curve: 
            - 동심타원(concentric ellipses)의 중앙(X점)에 가까울수록 variance가 작음. 
            - X는 attainable set 안에 있을수도, 밖에 있을수도 있음. 
        - 접점 $ \^{X}(E) $
            - isomean line과 isovariance curve의 접점
            - E(isomean line)이 더 커지면, isovariance curve와의 접점은 path trace를 남기게 된다. 
                - 궤적은 (처음엔) 직선으로 남음 (=critical line `l`)
                - 경제학의 indifference curve와 budget constraint line이 만나는 점의 궤적인 income-consumption curve 랑 비슷한거 같음. (이건 직선 아닐 수 있음)
        - 굵은선 efficient portfolio
            - 처음엔 critical line이 직선으로 접점의 궤도를 보이다가 attainable set의 경계에 닿으면 꺾여서 경계선 그대로 선택하게 됨. 
            - 그림 상 알 수 있듯, `X`점이 minimum variance고 `b`일 때가 maximum E 임. 
- 특이한 경우의 그래프
    - 
        - minimum variance point인 `X`가 attainable set 밖에 있을 때 
    - 
        - security가 3개일 때 
- 결론:
    - 대부분 일반적인 경우 diversified portfolio가 superior하다. 
    - 분산시켜도, 자산들의 correlation이 작아야 의미가 있다. 
        - rail companies 들만 가지고 아무리 분산해봐야 의미 없음. 
    - 두 자산이 같은 variance가지면 섞었을 때 variance 더 낮아짐. (diversification이 좋다.)
